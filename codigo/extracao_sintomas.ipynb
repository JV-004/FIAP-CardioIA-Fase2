# CardioIA – Parte 1: Extração de Sintomas e Diagnóstico

Este notebook lê frases de pacientes, identifica sintomas e sugere diagnósticos com base em um mapa de conhecimento.

```python
import pandas as pd

# 1. Leitura das frases de pacientes
with open('../dados/frases_sintomas.txt', 'r', encoding='utf-8') as f:
    frases = [linha.strip() for linha in f.readlines() if linha.strip()]

print(f'Total de frases carregadas: {len(frases)}\n')
for i, frase in enumerate(frases, 1):
    print(f'{i}. {frase}')
```

```python
# 2. Leitura do mapa de conhecimento (sintomas → doenças)
mapa = pd.read_csv('../dados/mapa_sintomas_doencas.csv')
print('Mapa de conhecimento carregado:')
mapa
```

```python
# 3. Função de extração de sintomas e sugestão de diagnóstico
def sugerir_diagnostico(frase, mapa):
    frase_lower = frase.lower()
    diagnosticos = []

    for _, row in mapa.iterrows():
        sintomas_linha = [str(row['sintoma1']), str(row['sintoma2']), str(row['sintoma3'])]
        sintomas_encontrados = [s for s in sintomas_linha if s.lower() in frase_lower]

        if sintomas_encontrados:
            diagnosticos.append({
                'doenca': row['doenca'],
                'sintomas_encontrados': sintomas_encontrados,
                'score': len(sintomas_encontrados)
            })

    if diagnosticos:
        melhor = max(diagnosticos, key=lambda x: x['score'])
        return melhor
    return None
```

```python
# 4. Processar todas as frases e exibir diagnósticos sugeridos
print('=' * 60)
print('RESULTADOS DO DIAGNÓSTICO ASSISTIDO')
print('=' * 60)

for i, frase in enumerate(frases, 1):
    resultado = sugerir_diagnostico(frase, mapa)
    print(f'\nPaciente {i}:')
    print(f'  Frase: {frase}')
    if resultado:
        print(f'  Sintomas identificados: {", ".join(resultado["sintomas_encontrados"])}')
        print(f'  Diagnóstico sugerido: {resultado["doenca"]}')
    else:
        print('  Nenhum sintoma reconhecido no mapa de conhecimento.')
```
